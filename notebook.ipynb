```python
# Retrieve the path to the current folders
current_path = os.getcwd() 

# Get the path to the csv file folder - in this case the 'data' file
csv_path = os.path.join(current_path, 'data') 

# Retrieve all csv files
all_csvs = os.listdir(csv_path) 

# Get the path to the location of each csv file
all_files = [os.path.join(csv_path, csv_file) for csv_file in all_csvs] 

# Allows you to import all csv files, each in a respective dataframe
all_dataframes = [] # Put the df in the form of a list

# Using this for loop and the zip function we iterate over all the csvs in the data
# The split method allows you to remove the ".csv", and it returns a list containing the following elements: ['the name of the excel file', '.csv']
# Thus, file[0] represents the name of the excel file since it is the first element in our tuple ['the name of the excel file', '.csv']

# ->>>>> WE SEARCH TO assigns each csv file name to its respective DF, e.g. helio.csv = helio (df) <<<<<-

for file in zip(all_files, all_csvs): 

    exchange_value = file[1].split('.')[0]
    df = pd.read_csv(file[0]) 
    df = df.assign(Exchange=exchange_value) 
        all_dataframes.append(df) 
```
